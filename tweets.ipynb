```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(r'C:\Users\SONY\Desktop\tweet\FinalBalancedDataset.csv\FinalBalancedDataset.csv')

# Print the column names
print(df.columns)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(r'C:\Users\SONY\Desktop\tweet\FinalBalancedDataset.csv\FinalBalancedDataset.csv')

# Print the column names
print(df.columns)

# Extract the tweet text and labels from the DataFrame
tweet_text = df['tweet']
labels = df['Toxicity']

# Bag of Words
vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(tweet_text)

# TF-IDF
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(tweet_text)

# Split the data into training and testing sets
X_train_bow, X_test_bow, y_train, y_test = train_test_split(X_bow, labels, test_size=0.2, random_state=42)
X_train_tfidf, X_test_tfidf, _, _ = train_test_split(X_tfidf, labels, test_size=0.2, random_state=42)

# Create an instance of Logistic Regression
model_bow = LogisticRegression()
model_tfidf = LogisticRegression()

# Train the models
model_bow.fit(X_train_bow, y_train)
model_tfidf.fit(X_train_tfidf, y_train)

# Evaluate the models
accuracy_bow = model_bow.score(X_test_bow, y_test)
accuracy_tfidf = model_tfidf.score(X_test_tfidf, y_test)

# Print the accuracies
print("Bag of Words Accuracy:", accuracy_bow)
print("TF-IDF Accuracy:", accuracy_tfidf)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Split the data into training and testing sets for Bag of Words representation
X_train_bow, X_test_bow, y_train, y_test = train_test_split(X_bow, df['Toxicity'], test_size=0.2, random_state=42)

# Create an instance of Decision Tree Classifier for Bag of Words representation
dt_bow = DecisionTreeClassifier()

# Train the model
dt_bow.fit(X_train_bow, y_train)

# Make predictions on the test set for Bag of Words representation
predictions_bow = dt_bow.predict(X_test_bow)

# Calculate accuracy for Bag of Words representation
accuracy_bow = accuracy_score(y_test, predictions_bow)

# Split the data into training and testing sets for TF-IDF representation
X_train_tfidf, X_test_tfidf, _, _ = train_test_split(X_tfidf, df['Toxicity'], test_size=0.2, random_state=42)

# Create an instance of Decision Tree Classifier for TF-IDF representation
dt_tfidf = DecisionTreeClassifier()

# Train the model
dt_tfidf.fit(X_train_tfidf, y_train)

# Make predictions on the test set for TF-IDF representation
predictions_tfidf = dt_tfidf.predict(X_test_tfidf)

# Calculate accuracy for TF-IDF representation
accuracy_tfidf = accuracy_score(y_test, predictions_tfidf)

# Print the accuracies
print("Bag of Words Accuracy:", accuracy_bow)
print("TF-IDF Accuracy:", accuracy_tfidf)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

# Split the data into training and testing sets for Bag of Words representation
X_train_bow, X_test_bow, y_train, y_test = train_test_split(X_bow, df['Toxicity'], test_size=0.2, random_state=42)

# Create an instance of Decision Tree Classifier for Bag of Words representation
dt_bow = DecisionTreeClassifier()

# Train the model
dt_bow.fit(X_train_bow, y_train)

# Make predictions on the test set for Bag of Words representation
predictions_bow = dt_bow.predict(X_test_bow)

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, predictions_bow)
recall = recall_score(y_test, predictions_bow)
f1 = f1_score(y_test, predictions_bow)

# Print precision, recall, and F1-score
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)

# Calculate confusion matrix
cm = confusion_matrix(y_test, predictions_bow)

# Print confusion matrix
print("Confusion Matrix:")
print(cm)

# Calculate ROC-AUC score
y_scores = dt_bow.predict_proba(X_test_bow)[:, 1]
roc_auc = roc_auc_score(y_test, y_scores)

# Plot ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
plt.plot(fpr, tpr, label='Decision Tree (AUC = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```
